```python
#creating new test df with specific columns
new_df={'host_id':df['host_id'],'host_name':df['host_name'],'neighbourhood_group' : df['neighbourhood_group'],'neighbourhood':df['neighbourhood'],'latitude':df['latitude'],'longitude':df['longitude'],'room_type':df['room_type'],'price': df['price'],'minimum_nights':df['minimum_nights'],'reviews_per_month': df['reviews_per_month'],'calculated_host_listings_count':df['calculated_host_listings_count']       }
ny=pd.DataFrame(new_df)
```

```python
'''Our 1st objective is to clean the data
since only the price column is in out control, we'll try detecting and tacling outliers from the column.'''
ny.describe()
```

 ## Removing outliers in price column



```python
#Removing Outliers in Price column
Q1 = np.percentile(df['price'],25)
Q2 = np.percentile(df['price'],50)
Q3 = np.percentile(df['price'],75)
IQR =  Q3 - Q1
a=Q1-(1.5*IQR)
b=Q3+(1.5*IQR)
ny1=ny[(ny['price']<b) & (ny['price']>a)]
```

```python
# ny1 is dataset containing the 'price' column

# Set the style of the plot
sns.set(style="whitegrid")

# Set the color palette
color_palette = ["#FFC300", "#FF5733", "#C70039", "#900C3F", "#581845"]

# Create the boxplot
ax = sns.boxplot(x=ny1['price'], color='#FFC300')

# Customize the appearance
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')

# Set the title and labels
ax.set_title('Boxplot of Price', fontsize=16, fontweight='bold')
ax.set_xlabel('Price', fontsize=12)
ax.set_ylabel('')

# Set the y-axis ticks
ax.yaxis.set_tick_params(color='gray')
ax.tick_params(axis='y', colors='gray')

# Set the x-axis ticks
ax.xaxis.set_tick_params(color='gray')
ax.tick_params(axis='x', colors='gray')

# Remove the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set the color of the whiskers and caps
for whisker in ax.artists:
    whisker.set_edgecolor('gray')

# Set the color of the medians
for line in ax.lines:
    line.set_color('gray')

# Customize the color of the boxes
for patch, color in zip(ax.artists, color_palette):
    patch.set_facecolor(color)

# Show the plot
plt.show()
```

# **Statistics: Mean, Median, Mode.**

Mean, median, and mode are three kinds of "averages". There are many "averages" in statistics, but these are, I think, the three most common, and are certainly the three you are most likely to encounter in your pre-statistics courses, if the topic comes up at all.

1.  The "**mean**" is the "average" you're used to, where you add up all the numbers and then divide by the number of numbers.

2.  The "**median**" is the "middle" value in the list of numbers. ## **Calculating The Difference Of Traffic Among Different Areas**

Except for Manhattan, in all the neighbourhood groups, 'Entire home/apt' is the room type that gets most traffic, although it is to most costlier type of property.

If we calculate with respect to neighbourhood we can see that none of top 10 locations have reviews for shared room but both the parameters give Allerton neighbourhood in Bronx with 'Entire home/apt' roomtype as the highest.

 Overall if we see, Allerton, Bronx has the highest traffic. We can conclude that throughout NewYork city there is a great preference towards Entire apt as compared to private or shared rooms irrespective of the prices hence greater is the traffic.
